```python
from scipy.sparse import coo_array
```

```python


def npz_to_coo_array(data, prefix="attr"):
    """convert data from npz file into coo sparse array
    convert data from https://github.com/shchur/gnn-benchmark/tree/master/data
      into sparse matrix """
    assert prefix in ["adj", "attr"]
    def to_indices(data):
        n = len(data[prefix+"_indptr"]) -1
        return np.repeat(np.arange(n), np.diff(data[prefix+"_indptr"]))
    
    return coo_array((data[prefix+"_data"], (to_indices(data), data[prefix+"_indices"])), shape=data[prefix+"_shape"])
```

```python
attr_array = npz_to_coo_array(data)
```

```python
npz_to_coo_array(data, "adj")
```

# (Base) Features

### Counting
Aggregations of the all ones vector
- Sum (e.g. degree is the Sum-Aggregated 1 neighborhood of count)
- Sum-Max (e.g. maximum degree of neighbors)
- Sum-Sum (total degree of neighbors)
- Sum-Threshold (number of neighbors with degree at least 3)

Aggregations of a Binary vector
- Sum (Number of green neighbors of a node)
- Sum-Max (neighborhood with the most green nodes)
- Max (Indicator whether there is at least one green neighbor)
- Min (Indicator whether all neighbors are green)
- Sum-Threshold (number of neighbors with at most three green neighbors)


Combinations/Readout
- Sum of binary over sum of all ones -> percentage of green nodes
e.g. the all ones vector, number of triangles, clustering coefficient,
