# Try to reproduce the word vectors attached to the "new" CORA dataset

```python
# function performin basic text preprocessing

import nltk
import string
nltk.download('stopwords')
from nltk.stem.porter import *
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

my_stopwords = set(["this", "use","learning", "paper"])

def process_abstracts(df, column, alt_column):
    en_stopwords = set(stopwords.words('english')) | set(string.punctuation) | my_stopwords
    vals = df[column]
    alt_vals = df[alt_column]
    stemmer = PorterStemmer()
    all_tokens = []
    for val, alt_val in zip(vals, alt_vals):
        if val is None:
            all_tokens.append(None)
        else:
            tokens = [token.lower() for token in word_tokenize(val) if token.lower() not in en_stopwords]
            stemmed_tokens = [stemmer.stem(token) for token in tokens]
            all_tokens.append([token for token in stemmed_tokens if token not in en_stopwords])
    return all_tokens
    
```

```python
df_token = df_to_iter[~df_to_iter["category_new"].isnull()]

tokens = process_abstracts(df_token, "Abstract:", "Title:")
token_dict = {key: value for key, value in zip(df_token["int_id"], tokens)}
```

### Calculate term frequency in new CORA

```python
df_new.drop([0,1434], axis=1).sum(axis=0)
```

```python
document_frequency = df_new.drop([0, 1434], axis=1).sum(axis=0)
```

```python
column_list = [None]*df_new.shape[1]
column_list2 = [Counter() for _ in range(df_new.shape[1])]
for _, row in df_new.iterrows():
    the_id = row[0]
    if token_dict[the_id] is None:
        continue
    the_tokens = set(token_dict[the_id])
    for column in np.nonzero(row)[0][1:]:
        if column_list[column] is None:
            column_list[column] = the_tokens
        else:
            column_list[column] &= the_tokens
            column_list2[column].update(the_tokens)
column_list2 = [col.most_common(3) for col in column_list2]
```

```python
column_list2[:20]
```

```python
column_list[:20]
```

```python
t1 = 0.7
t2 = 0.8
out = []
for suggestion, df in zip(column_list2[1:], document_frequency):
    if len(suggestion) < 2:
        out.append(None)
        continue
    word1, count1 = suggestion[0]
    _, count2 = suggestion[1]
    if  t2 > count2 / count1 and count1 / df > t1:
        out.append(word1)
    else:
        out.append(None)
    
```
